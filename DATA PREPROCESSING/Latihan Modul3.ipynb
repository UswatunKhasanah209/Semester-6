1. Import Library yang diperlukan

```python
# Tulis kode disini
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn import preprocessing
```

2. Tampilkan seluruh nilai dari dataset

```python
# Tulis kode disini
df = pd.read_csv('/content/drive/MyDrive/Penggalian Data/Modul3/Iklan_sosmed.csv')
display(df)
```

3. Lakukan EDA secukupnya

```python
# Tulis kode disini
print(df.info())
```

4. Cek apakah ada nilai yang hilang pada dataset dan isi nilai yang hilang pada 'EstimasiGaji' dengan median dari kolom tersebut.

```python
# Tulis kode disini
print(df['EstimasiGaji'].isnull().sum())
median_gaji = df['EstimasiGaji'].median()
df['EstimasiGaji'].fillna(median_gaji, inplace=True)
```

5. Lakukan encoding pada kolom 'Kelamin'

```python
# Tulis kode disini

label_encoder = preprocessing.LabelEncoder()

df['Kelamin']= label_encoder.fit_transform(df['Kelamin'])
df['Kelamin'].unique()
```

6. Standarisasi kolom 'Usia' dan 'EstimasiGaji' menggunakan StandardScaler.

```python
# Tulis kode disini
scaler = StandardScaler()

df[['Usia', 'EstimasiGaji']] = scaler.fit_transform(df[['Usia', 'EstimasiGaji']])
print(df)
```

7. Urutkan dataset berdasarkan 'EstimasiGaji' dari yang terendah ke tertinggi.

```python
# Tulis kode disini

df = df.sort_values(by='EstimasiGaji')
print(df)

```

8. Simpan dataset yang telah dibersihkan dan diproses ke file baru bernama 'Iklan_sosmed_clean.csv'

```python
# Tulis kode disini
df.to_csv('/content/drive/MyDrive/Penggalian Data/Modul3/Iklan_sosmed_clean.csv', index=False)
```
