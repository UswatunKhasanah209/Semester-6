```python
import pandas as pd

df = pd.read_csv("/content/drive/MyDrive/Penggalian Data/Modul2/smartphones.csv")

print("Metadata dataset: ")
print(df.info())

# 1. Apa model smartphone brand apple yang paling besar ramnya?
brand_apple = df[df['brand_name'] == 'apple']
max_ram = brand_apple.loc[brand_apple['ram_capacity'].idxmax()]['model']
print("\nModel smartphone Apple dengan RAM terbesar adalah:", max_ram)

# 2. Brand smartphone apa yang paling besar kapasitas baterai?
max_battery = df.loc[df['battery_capacity'].idxmax()]['brand_name']
print("Brand smartphone dengan kapasitas baterai terbesar adalah:", max_battery)

# 3. Brand smartphone apa yang rata-rata rating paling besar?
avg_rating = df.groupby('brand_name')['avg_rating'].mean()
max_rating = avg_rating.idxmax()
print("Brand smartphone dengan rata-rata rating terbesar adalah:", max_rating)

# 4. Model xiaomi apa yang kecepatan processornya paling cepat?
xiaomi_df = df[df['brand_name'] == 'xiaomi']
max_speed_xiaomi_model = xiaomi_df.loc[xiaomi_df['processor_speed'].idxmax()]['model']
print("Model smartphone Xiaomi dengan kecepatan processornya tercepat adalah:", max_speed_xiaomi_model)

# 5. Sebutkan model Brand smartphone Samsung apa saja yang internal_memorinya 128?
samsung_df = df[df['brand_name'] == 'samsung']
samsung_128 = samsung_df[samsung_df['internal_memory'] == 128]
samsung_128_models = samsung_128['model'].tolist()
print("Model smartphone Samsung dengan internal_memory 128GB adalah:", samsung_128_models)

print("\nAnalisis Deskriptif Statistik:")
display(df.describe())

Q1 = df.quantile(0.25)
Q3 = df.quantile(0.75)
IQR = Q3 - Q1

outliers = ((df < (Q1 - 1.5 * IQR) ) | (df > (Q3 + 1.5 * IQR))).sum()
display("Indeks outliers:", outliers)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('/content/drive/MyDrive/Penggalian Data/Modul2/smartphones.csv')

plt.figure(figsize=(10, 6))
avg_rating_price = df.groupby('brand_name')[['avg_rating', 'price']].mean().sort_values(by='avg_rating', ascending=False)
avg_rating_price.plot(kind='bar')
plt.title('Perbandingan rata-rata rating dan harga berdasarkan Brand')
plt.xlabel('Brand')
plt.ylabel('Rata-rata')
plt.xticks(rotation=90)
plt.show()

resolution_data = df.groupby('brand_name')[['resolution_height', 'resolution_width']].mean()
resolution_data.plot(kind='bar', figsize=(10, 6))
plt.title('Perbandingan Resolusi Tinggi dan Resolusi Lebar berdasarkan Brand')
plt.xlabel('Brand')
plt.ylabel('Rata-rata')
plt.xticks(rotation=90)
plt.show()

apple_df = df[df['brand_name'] == 'apple']
plt.figure(figsize=(10, 6))
plt.scatter(apple_df['price'], apple_df['internal_memory'], color='purple')
plt.title('Hubungan Harga dan Memori Internal pada Brand Apple')
plt.xlabel('Harga')
plt.ylabel('Memori Internal')
plt.show()
```
