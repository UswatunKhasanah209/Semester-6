```python
#Tugas1
import random

def Menambahkan(input_list):
    for _ in range(4):
        input_list.append(random.randint(1, 50))

def Menghapus(input_list):
    input_list = [input_list[i] for i in range(len(input_list)) if i % 2 == 0]
    return input_list

def Membalikkan(input_list):
    input_list.reverse()
    return input_list

def Menduplikasi(input_list):
    elemen_pertama = input_list[0]
    input_list.append(elemen_pertama)
    return input_list

def Menghilangkan(input_list):
    nilai_minimum = min(input_list)
    input_list.remove(nilai_minimum)
    return input_list

def Mengurutkan(input_list):
    input_list.sort()
    return input_list

def tampilkan_menu():
    print("Menu:")
    print("1. Menambahkan 4 elemen numerik acak.")
    print("2. Menghapus elemen di indeks ganjil.")
    print("3. Membalik urutan elemen dalam list.")
    print("4. Menggandakan elemen pertama dan menambahkannya ke akhir list.")
    print("5. Mencari dan menghapus elemen dengan nilai terkecil dalam list.")
    print("6. Mengurutkan list secara ascending.")
    print("7. Exit")

def output():
    my_list = [5, 3, 8, 2, 7]

    while True:
        print("\nList Asli:", my_list)
        tampilkan_menu()
        pilihan = input("Masukkan pilihan Anda (1-7): ")

        if pilihan == '1':
            Menambahkan(my_list)
        elif pilihan == '2':
            my_list = Menghapus(my_list)
        elif pilihan == '3':
            my_list = Membalikkan(my_list)
        elif pilihan == '4':
            my_list = Menduplikasi(my_list)
        elif pilihan == '5':
            my_list = Menghilangkan(my_list)
        elif pilihan == '6':
            my_list = Mengurutkan(my_list)
        elif pilihan == '7':
            print("Exit")
            break
        else:
            print("Pilihan tidak valid. Harap masukkan angka dari 1 hingga 7.")

if __name__ == "__main__":
    output()
```

```python
#Tugas2
def preprocessing_scores(scores):
    scores = [(subject, score) for subject, score in scores if subject != '' and score != '']

    max_scores = {}
    for subject, score in scores:
        if subject not in max_scores or int(score) > int(max_scores[subject]):
            max_scores[subject] = score

    filtered_scores = [(subject, max_score) for subject, max_score in max_scores.items()]

    return filtered_scores

def Menghitung(scores):
    return len(scores)

def Mengurutkan(scores):
    scores.sort(key=lambda x: int(x[1]))
    return scores

scores = [('Matematika', '85'),
          ('Biologi', ''),
          ('Kimia', '90'),
          ('Fisika', '80'),
          ('Matematika', '88'),
          ('Biologi', '95'),
          ('', '100'),
          ('Fisika', '82')]

Menyaring_scores = preprocessing_scores(scores)
print("Hasil setelah preprocessing:", Menyaring_scores)

total_entries = Menghitung(Menyaring_scores)
print("Jumlah total entri yang tersisa:", total_entries)

Mengurutkan_scores = Mengurutkan(Menyaring_scores)
print("Hasil pengurutan entri berdasarkan skor dari yang terendah hingga tertinggi:", Mengurutkan_scores)
```

```python
import numpy as np

array1 = np.random.randint(1, 30, size=8)
print("Array awal:")
print(array1)

minsc = np.min(array1)
maxsc = np.max(array1)
normalisasi1 = (array1 - minsc) / (maxsc - minsc)

print("\nHasil normalisasi array menggunakan Min-Max Scaling:")
print(normalisasi1)

normalisasi2 = np.sort(normalisasi1)[::-1]
print("\nArray setelah diurutkan dari nilai terbesar ke terkecil setelah dinormalisasi:")
print(normalisasi2)
```

```python
import pandas as pd

data = {
    'Merk Mobil': ['Toyota', 'Honda', 'Suzuki', 'Hyundai', 'Mazda', 'Ford', 'BMW', 'Mercedes', 'Mitsubishi', 'Nissan'],
    'Kota': ['Jakarta', 'Bandung', 'Jakarta', 'Surabaya', 'Bandung', 'Jakarta', 'Surabaya', 'Jakarta', 'Bandung', 'Surabaya'],
    'Harga (Juta Rupiah)': [200, 150, 120, 180, 220, 250, 500, 550, 200, 170],
    'Tahun': [2015, 2016, 2017, 2018, 2015, 2016, 2017, 2018, 2015, 2016],
    'Status': ['Sold', 'Sold', 'Available', 'Sold', 'Available', 'Sold', 'Sold', 'Available', 'Sold', 'Available']
}

df = pd.DataFrame(data)
print(df)

summary_stats = df['Harga (Juta Rupiah)'].describe()
print("\n=======================================================================\n")
print(summary_stats)

top_cities = df[df['Status'] == 'Sold']['Kota'].value_counts().nlargest(3)
print("\n=======================================================================")
print("\nTop 3 Kota dengan jumlah mobil terjual terbanyak:")
print(top_cities)

same_price_cars = df[df.duplicated(subset=['Harga (Juta Rupiah)'], keep=False)]
print("\n=======================================================================")
print("\nMobil yang memiliki harga yang sama:")
print(same_price_cars)

quartiles = df['Harga (Juta Rupiah)'].quantile([0.25, 0.5, 0.75])
def categorize_price(price):
    if price <= quartiles[0.25]:
        return 'Murah'
    elif price >= quartiles[0.75]:
        return 'Mahal'
    else:
        return 'Sedang'

df['Kategori'] = df['Harga (Juta Rupiah)'].apply(categorize_price)
print("\n=======================================================================")
print(df)
```
